```python
import matplotlib.pylab as plt
import myutils


model_type = "rnn"
model = myutils.get_model(model_type = model_type, num_classes = 2)
model.eval();
```

```python
import torch
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

path2weights = "./models/weights_"+model_type+".pt"
 
model.load_state_dict(torch.load(path2weights))
model.to(device);
```

```python
path2vido = "./data/ECHOS/class1/1.2.840.4267.32.100205641784865286778089171315423724299_1.2.840.4267.32.13847116838378289071976982464795994204.avi"
frames, v_len = myutils.get_frames(path2vido, n_frames=16)
len(frames), v_len
```

```python
#%matplotlib inline
import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
plt.figure(figsize=(10,10))
for ii,img in enumerate(frames[::4]):
    plt.subplot(2,2,ii+1)
    plt.imshow(img)
```

```python
imgs_tensor = myutils.transform_frames(frames, model_type)
print(imgs_tensor.shape, torch.min(imgs_tensor), torch.max(imgs_tensor))
```

```python
with torch.no_grad():
    out = model(imgs_tensor.to(device)).cpu()
    print(out.shape)
    pred = torch.argmax(out).item()
    print(pred)
```
